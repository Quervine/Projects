Radioactive decay and its uses

Radioactive decay is important to our health and many other things like mathematics and chemistry. We use radioactive decay to study living organisms, to diagnose and treat diseases, to sterilize medical equipment and food, to produce energy for heat and electric power, and to monitor various steps in all types of industrial activities.  Radioactive decay is used In the health industry, one way, is to help with food. We use radioactive emissions to expose food to the radiation to clean the food that we give patients so they can reduce the assumption of diseases from contamination from our food.   Radioisotopes have been used extensively in diagnosis and therapy. Radioactive isotopes are very good because they can be used as tracers and certain procedures. Since radioactive isotopes are identical chemically with stable isotopes of similar elements, they can take the place of those elements in certain medical processes. Not only that but because of their radioactivity they are very easy to trace.  Although many radioisotopes are used in tracing processes the most important seem to be iodine-131 phosphorus-32, and technetium-99m. Physicians use iodine-131 to determine things like cardiac output, plasma volume, and fat metabolism also to measure the activity of the thyroid gland where the isotope accumulates. Phosphorus-32 is useful in the identification of tumors because cancerous cells tend to accumulate phosphates more than normal cells do, which seems pretty intriguing. Technetium-99m, used along with radiographic scanning devices, is valuable for studying the anatomic structure of organs. There are also radioisotopes as cobalt-60 and cesium-137 that are widely used to treat cancer. They can be used selectively on tumors and minimize damage to adjacent healthy tissue.  

Applications of radioisotopes In industry are numerous. Many types of thickness gauges exploit the idea that gamma rays are weakened when they pass through material. By measuring the number of gamma rays, the thickness can be determined. This process is used in the automobile industry to test steel quality in manufacturing cars and to obtain the proper thickness of tin and aluminum. It is also used in aircraft industry to check for flaws in jet engines. In construction it is used to gauge the density of road surfaces and sub surfaces. In pipeline companies it is used to test the strength of welds. Oil, gas and mining companies use it to map the contours of test wells and mine bores and so on, the uses to radioisotopes in industry are also fairly endless. 

Radiometric dating is used in earth science to determine the age of rocks, Particular radioisotopes in geological material decays at a constant known rate to daughter isotopes. Using these techniques scientists have been able to determine the age of various rocks and rock formations. A special application, carbon-14 dating, have been useful to physical anthropologists and archaeologists. It helps them better organize the sequence of past events by enabling them to date more accurately fossils and artifacts from 500 to 50,000 years old.

How Radioactive decay happens

A nucleus is known as the core of an atom, made up of protons and neutrons, together those two particles take up the majority of the atoms mass, the main portion of the weight is in the core, the stability of this core is how isotopes and radioactive decay begin to formulate in an atom. In the atoms process of seeking stability, it will begin to decay emitting particles known as alpha, beta, positrons and gammas, which determine the type of radiation.
There are multiple ways that a nucleus can decay. Letâ€™s assume we have an atom composed of 8 protons and 9 neutrons. If we decide to emit two neutrons and protons from this nucleus, and this process is known as alpha decay, the emission of an alpha particles consists of removing neutrons and protons from the nucleus of the atom. When protons are removed, or decreased in numbers from the atom the element will change. Alpha decay is known as the process of a nucleus spitting out 2 neutrons and 2 protons in order to "slim down", apparently atoms dont like themselves very much and will do what they can to become stable. The emission of these 2 particles is what is known as alpha decay.

In our example the emission of two neutron and protons is known as a helium ion with a plus charge of 2, this is known as alpha decay, due to the lack of electrons.
If we take the same nucleus but now we focus on neutrons and the neutrons have a desire to become protons, hypothetically, they do this by emitting the electron particle, and this is known as beta decay, and a beta particle is just an emitted electron particle, since we are increasing the number of protons we are creating another element.  Now in the case of a proton wanting to become a neutron it will emit something called a positron which is the polar opposite of an electron, mass basically 0amu and charge plus 1amu and in this process called positron emission, this proton will become a neutron, and the atom will again change into a different element.
Gamma radiation is the emission of energy in the nucleus when the particles inside the nucleus rearrange themselves, the process of radioactive decay happens entirely at random, there is no way of telling which atoms will decay and which wont. When we continue to emit these particles we keep getting different elements, this is known as a decay chain(image by Tosaka)

Now all of these these processes are happening at a certain rate, but they are all different and not to mention all independent. Since decay is statistical and each decay is independent of each other, the bulk decay rate can be derived and it takes the form

$$\frac{dN}{dt} = -\lambda N$$

Where N is the population at a given time and lambda is the decay constant. The decay constant sets tine scales for the decay, basically the size of the decay will determine how fast or slow the decay will occure, if the decay is high then our population will decay at a very fastr rate, if our decay is slow our population will decay slowly. The uses for these two methods is very necessary. We can use fast decaying systems for patients with illnesses, what scientists do is they insert radioactive systems in patients and see results, and since they decay fast their is less risk of prolonged damage to the patients, just like the tracing we discussed earlier. The other form is used for stuff like determing the age of rocks, which we want to be as detailed as possible and since rocks arent humans we use the smallest decaying systems for this. I we wanted to figure out how much time it would take to decay a certain amount of atoms we would have to use a bit of math to do figure this out, we have to integrate our formula up above like so.

$$\frac{dN}{N} = -\lambda dt$$

$$\int\frac{dN}{N} = -\lambda \int dt$$

$$ ln(N) - ln(N_0) = -\lambda t$$

$$ ln(\frac{N}{N_0}) = -\lambda t$$

$$\frac{N}{N_0} = e^{-\lambda t}$$

$$N = N_0e^{-\lambda t}$$


Now we can input values into our equation and see what we get. The concept of half life is very well known and is directly related to our radioactive decay function. Half life is defined as, the amount of time it takes for a population to be half its original size, this is simple enough, the only difficult part really is determing what our $\lamba$ would be and what relationship it holds with our equation above, and to do that we need to use some values, now we will substitute t for the half life variable, we will use 50 for $N$ and 100 for $N_0$.

$$ 50 = 100e^{-\lambda \tau_{\frac{1}{2}}}$$

$$ \frac{50}{100} = e^{-\lambda \tau_{\frac{1}{2}}}$$

$$\frac{100}{50} = e^{\lambda \tau_{\frac{1}{2}}}$$

$$ln(2) = \lambda \tau_{\frac{1}{2}}$$

$$ \lambda = \frac{ln(2)}{\tau_\frac{1}{2}}$$

$$\tau_\frac{1}{2} = \frac{ln(2)}{\lambda}$$

Eulers method is a numerical tool for approximating values for solutions of differential equations. The eulor method is simple approach to solving differential equations that is easy to derive, lets take this for an example,

$$y'(t) = f(t, y(t)), \hspace{0.5cm} y(t=0) = y_0$$ 

Where $y'$ is the first derivative with respect to time, $y''$ is the second derivative with respect to time and so on. we can then turn this into another form of the equation if we know our values of $t_n$ and $t_{n+1}$, the derivative can be written as 

$$y'_n = \frac{y_{n+1} - y_n}{t_{n+1} - t_n}.$$

We can then write down the difference equation as 

$$\frac{y_{n+1} - y_n}{t_{n+1} - t_n} = f(t_n, y_n).$$

and finally the simplest form of this is

$$y_{n+1} = y_n + hf(t_n, y_n).$$

This equation lets us sort of look forward in time when we generalize the equation to 

$$y^{(N)}(t) = f(t, y(t), y^{(1)}(t), ... y^{(N-1)}(t).$$

The way to go about this is to convert this into a system of equations like so.

$$z_1(t) = y(t), z_2(t) = y^{(1)}(t), z_3(t) = y^{(2)}(t), ..., z_N(t) = y^{(N-1)}(t).$$

We now have a system of equations. If we consider the $z_i(t)$ to be a vector, we can write down 

$$\textbf{z} = \begin{pmatrix}
z_1(t) \\
z_2(t) \\
z_3(t) \\
.      \\
.      \\
.      \\
z_N(t)
\end{pmatrix}.$$

The system of equations then becomes, 

$$\textbf{z'}(t) = \begin{pmatrix}
z_2(t) \\
z_3(t) \\
z_4(t) \\
.      \\
.      \\
.      \\
f(t, z_1(t), ..., z_N(t)
\end{pmatrix}.$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def radioactive(N=0.1, lamb = 0.1):
    """
    Function returns the output -lamb*N. 
    
    Keywords:
    x: Input. This is the population of the species. Can be concentration too so it is ok to input floats.
    lamb: Input. This is the decay constant. Note that lambda in python has a special meaning. So we name
          this variable lamb.
    output: returns -lamb*N.
    
    """
    
    return -lamb*N
```

```python
def euler(func, h=0.1, yCurrent=0.1, *funcArgs):
    """
    Function performs one Euler step returns the value for the next iteration.
    
    Keywords:
    func: right hand side of the differential equation. 
          Default is the radioactivity function i.e. radioactive().

    h: Input. This is the time step used to the advance the solution to the next time point.
    yCurrent: Value of variable at current time step.
    *funcArgs: Parameters for the function in the differential equation.
    yNext: Output. Value of the variable at the next time step.
    
    """
    
    # Initialize
    yNext = 0.0
    
    # Calculate next time step
    yNext = h*func(yCurrent, *funcArgs) + yCurrent
    
    return yNext
```

Now we want to use our new found eulor method to calculate values and see if we get something we can qualify as viable.

```python
numIter = 10
y = np.zeros((numIter,1))
y[0] = 50.0
h = 0.5
lamb = 1.0
tList = np.linspace(0, h*numIter, numIter)

for i in range(1, len(y)):
    y[i] = euler(radioactive, h, y[i-1], lamb)
```

```python
fig, axes = plt.subplots(figsize=(12, 5))

axes.plot(tList, y,  '.', markersize=12)
axes.set_xlabel('time', fontsize=15)
axes.set_ylabel('y', fontsize=15)
axes.grid(True)
```

### Runge-Kutta Method

The Runge-Kutta methods are a family of explicit and implicit iteritive methods that are used for approximations of Ordinary Differential Equations, one being in conjunction with the euler method.  These methods where developed around the 1900 by german mathematicians Carl Runge and Martin Kutta. The RK4 method is the most widely used method and has an initial value problem of

$$ y' = f(t,y)$$

and 
$$y(t_0) = y_0$$

Where y is an unkown factor of time t, which is what we are trying to approximate. We define our method as follows


$$y_{n+1} = y_n + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4),$$

where,

$$k_1 = f(t_n, y_n),$$

$$k_2 = f(t_n + \frac{h}{2}, y_n + \frac{hk_1}{2}),$$

$$k_3 = f(t_n + \frac{h}{2}, y_n + \frac{hk_2}{2}),$$ and

$$k_4 = f(t_n + \frac{h}{2}, y_n + hk_3).$$


here $y_{n+1}$ is the RK4 approximation of $y(t_{n+1})$ and the next value of $y_{n+1}$ is determined by the value of $y_{n}$ and we then add the weighted average of the four increments 

* $k_1$ is based on the slope of the first point of y(Euler Method)
* $k_2$ is the increment based on the slope at the midpoint of the interval, y and k1
* $k_3$ is the same as above, except now we are using y and k2 instead of k1
* $k_4$ is the increment based on the last point using y and k3

After we average the four weights, greater weight is given to the points in the midterm. The RK4 method is a fourth order differential equation which means it has a truncation error of $O(h^5)$ while the total accumulated error is $O(h^4)$, this means our error is fairly small, espeically compared to the euler method.

The Explicit Runge-Kutta method is used for a wider varitey of weights and it follows as

$$y_{n+1} = y_n + h\sum{b_ik_i}$$ 

Where $i$ goes from $1$ to $s$.  Here

* $k_1 = f(t_n,y_n)$
* $k_2 = f(t_n + c_2h, y_n + h(a_{21}k_1))$
* $k_3 = f(t_n+c_3h, y_n + h(a_{31}k_1 + a_{32}k_2))$

* $k_s = f(t_n + c_sh, y_n +h(a_{s1}k_1+...+a_{s,s-1}k_{s-1})$

As you cn see the slopes we find for k1, ..., k4 are somehwat hugging the graph, but differently than euler. In euler we kind of venture a bit below the graph and keep approximating along that one line, but here we get four different slopes around our graph, some above some below one straight throught the middle and then we average them all together to get our approximation.

Lets make a numerical example before we get into the code so we can see this formula implemeneted and how the process looks by hand. 

$$\frac{dy}{dx} = x^2y$$

$$y(0) = 1$$

$$\triangle{x} = 0.1$$

$$k_1 = f(t_n,y_n) = 0^2 * 1 = 0$$

$$k_2 = f(t_n + \frac{\triangle{x}}{2}, y_n + \frac{k}{2})$$

$$k_2 = (0 + \frac{1}{2}*(0.1))^2 * (1 + \frac{1}{2}(0)(0.1)) = 0.0025$$

$$k_3 = f(x + \frac{\triangle{x}}{2}, y_n + \frac{k_2*\triangle{x}}{2})$$

$$k_3 = ((0 + \frac{(0.1)^2}{2})*(1+\frac{(0.0025)(0.1)}{2})) = 0.0025$$

$$k_4 = f(x + \triangle{x}, y_n + k_3 * \triangle{x})$$

$$k_4 = (0 + (0.1)^2) * (1 + 0.0025*0.1) = 0.01000$$

Now we want to pt everything together using our formula which will be something like,

$$y_{i+1} = y_i + \frac{\triangle}{6}(k_1 + 2k_2 + 2k_3 + k_4)$$

$$y(0.1) = 1 + \frac{0.1}{6}(0 + 2(0.0025) + 2(0.0025) + 0.001) = 1.00333$$

So basically what we are saying is thay when x is 0.1 y is 1.00333.

Now what we want to do is find the value of y whe x is 0.2, we could do 0.3, 0.4 and so forth...Lets see how close to 1 we can really get, we will only be doing one more example since they are a bit ardurous and lengthy but it should be more than enough to prove our point. 


$$k_1 = f(t_n,y_n) = (0.1)^2 * 1 = 0.0100$$

$$k_2 = f(t_n + \frac{\triangle{x}}{2}, y_n + \frac{k}{2})$$

$$k_2 = (0.1 + \frac{1}{2}*(0.1))^2 * (1.000333 + \frac{1}{2}(0.01)(0.1)) = 0.02252$$

$$k_3 = f(x + \frac{\triangle{x}}{2}, y_n + \frac{k_2*\triangle{x}}{2})$$

$$k_3 = ((0.1 + \frac{(0.1)^2}{2})*(1.00333+\frac{(0.02252)(0.1)}{2})) = 0.02253$$

$$k_4 = f(x + \triangle{x}, y_n + k_3 * \triangle{x})$$

Now we should expect to see a solution very close to our exact line graph when we graph our RK4 method, lets see how it looks.

```python
def rk4(func, h=0.1, yCurrent=0.1, *funcArgs):
    """
    Function performs one RK4 step returns the value for the next iteration.
    
    Keywords:
    func: right hand side of the differential equation. 
          Default is the radioactivity function i.e. radioactive().
    h: Input. This is the time step used to the advance the solution to the next time point.
    yCurrent: Value of variable at current time step.
    *funcArgs: Parameters for the function in the differential equation.
        
    yNext: Output. Value of the variable at the next time step.
    
    """
    
    # Initialize
    yNext = 0.0
    
    # Calculate k1, k2, k3 and k4
    
    k1 = func(yCurrent, *funcArgs)
    k2 = func(yCurrent + h*k1/2, *funcArgs)
    k3 = func(yCurrent + h*k2/2, *funcArgs)
    k4 = func(yCurrent + h*k3, *funcArgs)
    
    # Calculate next time step
    yNext = (h/6)*(k1 + 2*k2 + 2*k3 + k4) + yCurrent
    
    return yNext
```

```python
numIter = 50
yRK4 = np.zeros((numIter,1))
yRK4[0] = 50.0
h = 0.5
lamb = 1.0
tList = np.linspace(0, h*numIter, numIter)

for i in range(1, len(yRK4)):
    yRK4[i] = rk4(radioactive, h, yRK4[i-1], lamb)
```

```python
fig, axes = plt.subplots(figsize=(15, 12))

axes.plot(tList, yRK4, '.-', markersize=8)
axes.plot(tList, y[0]*np.exp(-lamb*tList), '-', label='Exact solution')
axes.set_xlabel('time', fontsize=15)
axes.set_ylabel('y', fontsize=15)
axes.grid(True)
plt.xlim(0,10)
```

You can still take the Rk4 method much further than the euler method which is a good advantage.  The euler method itself has a set of disadvantages that should be discussed.  The euler method is very easy to use but it has many errors that accumualate over time from iteration to iteration. If you use a large h then you will see that the method becomes too innacurate and we classify these as truncation errors. Our series that we ue for euler is a taylr expansion of the form

$$y_{n+1} = y_n + h(ft_n, y_n).$$

This first order equation is what we use to approximate our answer in every iteration, not the cleanest way to do it but it gets you a rough answer.
```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

a = 0.2
b = 0.2
c = 2.5

def Rossler(A, t):
    x, y, z = A  # unpack the state vector
    return  -y - z, x + a * y, b + z * (x - c)  # derivatives
    """
        Our function Rossler is meant to take our values for the rossler equation
       and vectorize them for later use.
       
       
    """

A0 = [1, 1, 1]
time = np.arange(0, 20, .01)

A = odeint(Rossler, A0, time)

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(A[:,0], A[:,1])
plt.show()
```

```python
from numpy import *
from scipy import *
from pylab import figure
%matplotlib inline

def rossler_bifurc(x,y,z,h,a,b,c):
    x_n=x+h*(-y-z)
    y_n=y+h*(x+a*y)
    z_n=z+h*(b+z*(x-c))   
    return x_n,y_n,z_n


a=0.2
b=0.2
c=5.7

t_ini=0
t_fin=120
h=0.0001
timesteps=int((t_fin-t_ini)/h)
t=linspace(t_ini,t_fin,timesteps)

x=zeros(timesteps)
y=zeros(timesteps)
z=zeros(timesteps)

for k in range(x.size-1):
    [x[k+1],y[k+1],z[k+1]]=rossler_bifurc(x[k],y[k],z[k],t[k+1]-t[k],a,b,c)
```

```python
fig =figure(figsize=(22, 22))
ax1 = fig.add_axes([0.1, 0.7, 0.4, 0.2])


ax1.plot(t, x,color='red',label='x(t)')
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')
ax1.legend()
ax1.axis((t_ini,t_fin,min(x),max(x)))
```
