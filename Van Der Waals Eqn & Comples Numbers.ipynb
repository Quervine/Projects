# Part I - Critical Points of the Van Der Waals Equation

The main idea behind this formula is to show that when P and V are multiplied they form some constant number. Van Der Waal was stating that if real gasses are to exist in nature they have to break the two rules of ideal gasses, they must have intermolecular forces, which change when you change P,V and T, and gas molecules have volume. We  have to account for the part of the equation that breaks down, so we work with the reduced gas equation
One writes the equation in reduced form, by making the following variable adjustments,

$p_R = \frac{P}{P_C},$
$v_R = \frac{V}{V_C},$
$T_R = \frac{T}{T_C},$

where

$P_C = \frac{a'}{27b'^2},$ $V_C = 3b'$ and
$T_R = \frac{8a'}{27kb'}.$

giving us the reduced equation of state

$$(p_R + \frac{3}{V_R^2})(v_R - \frac{1}{3}) = \frac{8}{3}T_R.$$

The good thing about writing the equation in the reduced form is that regardless of the gas, the equation will still work, because we took out the constants $a$ and $b$ which are related to the specific gas properties.

The bisection method is a fairly simple root calculating method in which we take an interval between two points of interests, point a and point b we then bisect, or diessct, until we converge, meet, up to a certain point, which will be our root, now keep in mind we want three roots not just one. 

The algorithm is as follows:

 1. Choose $[a, b]$ interval of the root such that $sign(f(a)) \neq sign(f(b))$.
 2. Calculate $f$ at the midpoint of the bracket i.e. $c = (a + b)/2$.
 3. Replace $a$ or $b$ in such a way that the *new* $f(a)$ and $(f(b)$ also have opposite signs.
 4. Keep repeating steps 1-3 till a certain tolerance is reached or if the bracket width $|b - a|$ reaches a certain defined minimum.
 5. The final answer is the last value of $c$ which meets tolerance, our tolerance is simply the furthest we allow our answer to go beyond certain significant figures.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def myFunction(V=0.1):
    """
    Function returns the output cos(x) - x. The value of x should be in radians. 
    
    Keywords:
    x: Input. For the cosine function, this must be in radians. Default x = 0.1
    
    """
    T = 0.8
    P = 0.4
    return (8*T/(3*V-1))-(3/V**2)- P
```

```python
def bisectionFunc(f=myFunction, a=0.4, b=0.6, tolerance=1e-7):
    """
    Returns the root of the input function f and the number of iterations it takes as
    a pair using the bisection method.
    
    Keywords:
    f: function for which root is to be found. Default is myFunction
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7
    
    """
    # First figure out signs of a and b
    signa = np.sign(f(a))
    signb = np.sign(f(b))
    iter = 0
    
    # Write loop for the bisection method. Don't stop till tolerance is met
    while b-a > tolerance:
        c = (a+b)/2
        print("Best guess of root = {} at iteration {:3d}".format(c, iter))
        
        signc = np.sign(f(c))
        
        # What happens if c is zero? It means root is found!
        if signc == 0:
            return c
        elif signc == signa:
            a = c
        else:
            b = c
            
        iter += 1
    
    # Return final value
    return c, iter 
```

### Newton's method (also Newton-Raphson method)

The bisection method seems to work pretty good. But the function we used was a simple function. This function will get complicated but it will also increase runtime.

An alternative is a gradient based Newton method. The algorithm works like this

 1. The function must be differentiable as oppsed to the bisection method.
 2. First find the derivative of the function, the tangent line.
 3. Elongate the tangent line and calculate where it crosses the x-axis. That gives $x_{i+1}$ a new estimate of the root.
 4. Repeat till tolerance is achieved.
 
 The iteration formula is then
 
 $$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}.$$
 
 Now we need to Derive our function:
 
 $$\frac{8T_R}{3V_R-1} - \frac{3}{V^2_R} - P = 0$$
 
 Lets first input our values of $T_R$ and $V_R$ which will be $0.4$ and $0.8$ respectively. When we derive this equation, with respect to $V_R$ we get:
 
 
$$\frac{6}{V_R^3} - \frac{24T_R}{3V_R-1^2} = 0$$

Now that we have our function and the derivative we can input it into our function and we should see values similar to what we got for our Bisection method.

```python
def myFunction2(x=0.1):
    """
    Function returns the output cos(x) - x and its derivative as a pair of values. 
    The value of x should be in radians. 
    
    Keywords:
    x: Input. For the cosine function, this must be in radians. Default x = 0.1
    
    """
    T = 0.8
    P = 0.4
    # Return function value and derivative as a pair
    return (8*T/(3*x-1))-(3/x**2)- P, (6/x**3) - (24*T/((3*x-1)**2))
```

```python
def newtonMethod(fx = myFunction2, xInit = 0.5, tolerance = 1e-8):
    """
    Returns the root of the input function f and the number of iterations it takes as
    a pair using the Newton's method.
    
    Keywords:
    fx: function for which root is to be found. Default is myFunction2. fx must
       return both function value and derivative at x.
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7
    
    """
    
    # Initialize the variables
    step = 0
    f, df = fx(xInit)
    dx = -f/df
    x = xInit
    
    # Run the iterations in the algorithm
    while np.abs(dx) > tolerance:        
        print("Best guess of root = {} at iteration {:3d}".format(x, step))
        f, df = fx(x)
        dx = -f/df
        x += dx
        step += 1
    
    # Return output
    return x, step
```

If we go back to our previous equation 


$$ z^3 = 1$$  

we can begin to work on this and see why fractals are so interesting. First what we want to do it factor this equation to get its roots, now since it is a polynomial with an exponent power of three it will have three roots, but how do we find these roots? Well we can rewrite this to represent eulors formula like this 


$$ 1 = e^{0 i}$$ 

And this is correct mathematically because e to the power 0 is 1. Now If we think about this as a circle 0 is the starting point and we can circle back around with $2\pi$ and the same for $4\pi$,

$$e^{0 i} = e^{2\pi i} = e^{4\pi i}$$

So $$z^3 = e^{2\pi i} = e^{4\pi i}$$

Now if we recall a bit of differential equations we will notice something a bit intereting about eulors formula which is what we will use to solve for the roots of $Z^3$.


$$ e^{x i} = cos(x) + isin(x)$$

is the general form for eulors method and when we solve for $Z$ up above and input it into our formula we get


$$ e^{\frac{2\pi}{3} i} = cos(\frac{2\pi}{3}) + isin(\frac{2\pi}{3})$$

this is also equal to 

$$ e^{\frac{4\pi}{3} i} = cos(\frac{4\pi}{3}) + isin(\frac{4\pi}{3})$$

Now to solve for these we need a trigonometric table which i will show and then we can get our answers


![crit](https://www.cubety.tk/wp-content/uploads/4we2PR/omtex-classes-trigonometric-table.jpg)


Now its important to know that $\frac{2\pi}{3}$ is $60^\circ$. 

When we look through our table we get our answer for our roots and we get 


$$ Z = 1, Z = -\frac{1}{2} + \frac{\sqrt3}{2\pi} , Z = -\frac{1}{2} - \frac{\sqrt3}{2\pi}$$

or put more simply 

$$Z = [1+ 0i, -0.5+0.8660254i, -0.5-0.8660254i]$$

Now we want to use our Newton Raphson Method to get these same roots and plotting them to see what we get.

```python
def myFunction2(x=0.1):
    """
    Function returns the output cos(x) - x and its derivative as a pair of values. 
    The value of x should be in radians. 
    
    Keywords:
    x: Input. For the cosine function, this must be in radians. Default x = 0.1
    
    """
 
    # Return function value and derivative as a pair
    return x**3 - 1, 3*(x**2)
```

```python
def newtonMethod(fx = myFunction2, xInit = 0.5, tolerance = 1e-8):
    """
    Returns the root of the input function f and the number of iterations it takes as
    a pair using the Newton's method.
    
    Keywords:
    fx: function for which root is to be found. Default is myFunction2. fx must
       return both function value and derivative at x.
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7
    
    """
    
    # Initialize the variables
    step = 0
    f, df = fx(xInit)
    dx = -f/df
    x = xInit
    
    # Run the iterations in the algorithm
    while np.abs(dx) > tolerance:        
        print("Best guess of root = {} at iteration {:3d}".format(x, step))
        f, df = fx(x)
        dx = -f/df
        x += dx
        step += 1
    
    # Return output
    return x, step
```
