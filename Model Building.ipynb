All of the blue dots are seperate points that corespond to the same topic.  Now we can see the line of best fit is put out through the middle of the data points but there is a way to calculate how to make the best fit line.  A best fit line/curve is whatever line/curve best estimates the data we are presenting. Lets assume we have information regarding pairs of $(x_i, z_i)$ and lets say we have a function $$f(x, \beta) = \sum_{j = 1}^m \beta_j\phi_j(x),$$ 

here $\beta_j$ are the parameters which are used to determine the characterisitics of the curve. This function is supposed to predict valule of $z_i$ for a given $x_i$. $\phi_j$ are the functions used along side the parameters to fit the curve, which is linear for $\beta_i$.

If you notice our example up above you will see that there are many points very far away from the curve/line and we refer to these points as noise in the data. When calculating our curve of best fit we have to take into account this noise and we try to predict the values of $z_i$ as accurately as possible. The main objective is to find the best parameters $\beta_j$ which will result in the best approximation to the data we have using a chosen model. 


Defining the error in our model is the most optimal way of making the line of best fit as accurate as possible, to do this we will need the distance from $z_i$ and $f(x_i, \beta)$. We can write this as 

$$r_i = z_i - f(x_i, \beta).$$

We can define our error(Objective Function) as

$$S = \sum_{i=1}^nr_i^2.$$


The next step is to calculate the $\beta_j$ so the $S = S_{min}$. This can be done like this:



$$\frac{\partial S}{\partial \beta_j} = 2\sum_ir_i\frac{\partial r_i}{\partial \beta_j} = 0, j = 1, 2, .., m.$$

We know $r_i = y_i - f_i$, we can write down

$$-2\sum_ir_i\frac{\partial f(x_i, \beta)}{\partial \beta_j} = 0, j = 1, 2, ..., m.$$

The $\beta_i$ that will meet these conditions is the best fit parameters which will produce the best fit line/curve $f(x_i, \beta)$ This is known as minimizing the objective function.

If we write: 

$$X_{ij} = \frac{\partial f(x_i, \beta}{\partial \beta_j} = \phi_j(x_i),$$  we can manipulate this to get 

$$\hat\beta = (X^TX)^{-1}X^T\textbf{y}.$$

where $\hat\beta$ is the matrix containing $\beta_i , X$ is the $n x m$ matrix containig $X_ij$ and $y$ is the vector containing the lists of $y_i$ values, we will do this more in detail later for now lets begin with plotting our simple plots.

We are now asked to plot the graphs and use our curve fitting machinery for both cases. We are to get it from a .py file. In order to do this processs we need to create an array of our data and arange it because we will need to do matrix multiplication later on in order to calculate $\beta_0$ and $\beta_1$ Numpy has matrix multiplicatio routines built in. In the normal equations, the product is the common matrix product. we are looking to solve for the form 

$$Ax = B,$$ 

where $A, B$ and $x$ are matrices, we want to solve for $x$ and it will be written as 

$$x = A^{-1}B.$$
We do this by using the 'linalg.solve()' algorithm that is inputted in numpy already. and the syntax is written as 

$$x = np.linalg.solve(A, B)$$

```python
n = len(Steel)
heightSteel = np.array([Steel[:, 0]]).T
#print(heightSteel)
```

```python
timeSteel = np.array([Steel[:, 1]]).T
#print(timeSteel)
```

```python
time2Steel = timeSteel**2
#print(time2Steel)
```

```python
XSteel = np.array([np.ones(n), Steel[:,1]**2]).T
#print(XSteel)
```

```python
XPlastic = np.array([np.ones(n), Plastic[:,1]**2]).T
```

The way matrix multiplication works for finding beta is very simple we want to transpose our first argument and multiply it by our second argument and the result will be a beta function this image form wikipedia (by Tarquin) shows the process in a more visual manner.


![crit](https://upload.wikimedia.org/wikipedia/commons/2/26/Matrix_multiplication_diagram.PNG)

```python
betaSteel = np.linalg.solve(XSteel.T.dot(XSteel), XSteel.T.dot(heightSteel))
print(betaSteel)
```

We obtained the values $\beta_0 = -2.41302530e-01$ and $\beta_1 = 4.87245487e+02$ i.e. the slope and intercept of the best fitting straight line of our future curve fitting. Now we need to plot our graph and fit the curve.

```python
# plot the line to check that we get something reasonable
fig, ax = plt.subplots(figsize=(12,5))

ax = fig.add_axes([0,0,1,1])

ax.scatter(time2Steel, heightSteel, label="y vs t^2 Steel")

# Plot the fit
xx = np.linspace(0.04,0.10,10)
yy = betaSteel[0] + betaSteel[1]*xx

fitLabel = "linear fit, $y = \\beta_0 + \\beta_1 t^2$\n beta_0={}, beta_1={}".format(betaSteel[0], betaSteel[1])
ax.plot(xx, yy,  'r', label = fitLabel)
ax.legend(loc ='upper left')
```

We were already expecting our line to be a straight line and the points fit very well on our curve so it seems like a success. Our $\beta_0$ was very small at $-.24130253$ which is what we expected because the matrices that were being multiplied was time and it was squared as well, while our $\beta_1$ was fairly large at $487.245486551$ which also was what we expected becuase you were multiplying the height with the $t^2$ , we will do the same for our Plastic but we will use a function 'normalEquationSolver' to calculate our betas since it is much easier , and again we want to see similar values to our betas for plastic.

```python
def normalEquationSolver(X, height):
    """
    This function solves the normal equations for linear regression using
    least squares method by taking in the values of X and y. Here, y is the 
    target and X is constructed using the model as X_ij = phi_j(x_i). 
    The output is an array containing the best fit parameters for the model
    using the data.
    
    Keywords:
    X: this is the matrix containing the values (phi_j(x_i))
    y: this matrix contains the values of y which the model tries to predict
    beta: this is the output of the function. It contains the parameters
          of the fitting model in an array.
    
    """
    import numpy as np
    
    # Solve the matrix equation (X'X)beta = X'y for beta 
    # where ' means transpose
    
    beta = np.linalg.solve(X.T.dot(X), X.T.dot(heightSteel))
    
    return beta
```

```python
betaPlastic = normalEquationSolver(XPlastic,heightPlastic)
print(betaPlastic)
```

The equation for this is as follows
   
   
   $$y_{n+1} = by_n(1-y_n).$$

In the equation $y_n$ is the population of the species at the $n$th time iteration and $0 \leq b \leq 4$ is a parameter that controls the system behavior along the initial population $0 \leq Y_0 \leq 1$.  The fixed points of the equation are the equilibrium points of the equation. If the system is started a certain point it remains at that point. The reason why we want to show these points is to show what numbers for $b$ or fertility rate would be at a fixed $y$ so basically what fertility rate we would need to have to maintain a population of 5million constant.So to show this we want to use our mayfly model.

```python
# First import necessary modules
import numpy as np
import numpy.matlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mayfly(b=2, yInit=0.01, n=10):
    """
    Function returns the time series of population for n iterations as 
    a numpy array. 
    
    Keywords:
    b: parameter in the model
    yInit: initial population
    n: number of iterations to be performed
    out: Numpy array containing the populations at the n time iterations
    
    """
    # Initialization
    out = np.zeros((n,1))
    y = yInit
    
    # Set up loop for the iterations
    for i in range(n):
        out[i] = y
        y = b*(1-y)*y
        
    # Send back output and exit
    return out
```

```python
b = .5
yInit = 0.5
n = 50
pop = mayfly(b, yInit, n)
print(pop)
```

```python
# create vectorized mayfly model

def mayfly2(b = np.linspace(0, 4, 50),
            yInit=np.matlib.repmat(0.01, 50, 1),
           n=50):
        """
        creates a 2d array of populations, rows correspond to time instants
        and columns correspond to various values of b.
        
        Keyords: 
        b: numpy array containing many different parameter values
        yInit: initial conditions for all these b values
        n: number of iterations
        out: 2d numpy array containing populations for various values of b
        columns correspond to specific values of b and rows correspond
        to time instants
        
        
        """
        
        #initialize
        
        m = len(b)
        out = np.zeros((n,m))
        
        # We will transpose yInit to a cloumn vector
        
        y = yInit.T
        
        # Set up loop for iterations
        
        for i in range(n):
            out[i, :] = y
            y = b*(1-y)*y #broadcasting
        
        return out
```
